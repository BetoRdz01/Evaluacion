## **Limpieza de Valores Atípicos**

La identificación de datos atípicos se utilizará con rangos intercuartílicos dado a que de esta manera se pueden visualizar los datos adecuadamente así como también se podra ser más estricto con los rangos, obteniendo datos menos dispersos. 
## Cal

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=Calnl
percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25
Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
outliers_iqr= Calnl[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
#Obtenemos datos limpios del Dataframe
Calnl_iqr= Calnl[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
Calnl_iqr.isnull().sum()
```

```python
# Reemplazamos valores nulos (atípicos) con la mediana, siendo otra medida de tendencia central que nos ayudará 
# a tener los valores atípicos como valores con comportamiento central dentro de la distibución de los datos.
Calnl_iqr2=Calnl_iqr.copy()
Calnl_iqr2=Calnl_iqr2.fillna(round(Calnl_iqr.median(),1))
Calnl_iqr2.isnull().sum()
```

```python
Cal = pd.concat([Calol, Calnl_iqr2], axis=1)
Cal.to_csv('California_limpio.csv')
```
