```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Importamos Archivos
Cal = pd.read_csv('California_limpio.csv')
Mex = pd.read_csv('Mexico_limpio.csv')
Esp = pd.read_csv('Girona_limpio.csv', encoding='unicode_escape')
```

# host_response_time

```python
plt.figure(figsize=(15,5))
Esp['host_response_time'].value_counts()[Esp['host_response_time'].value_counts().values>1].plot(label='Girona', linestyle='dashed', color='#BB2222')
Mex['host_response_time'].value_counts()[Mex['host_response_time'].value_counts().values>1].plot(label='Ciudad de México',color='#22BB13')
Cal['host_response_time'].value_counts().plot(label='California', color='#22BBBB')
plt.legend()
plt.title('HOST RESPONSE TIME POR CIUDAD')
plt.ylabel('Frecuncia')
plt.show()
```

Se obsera como Ciudad de México tiene un gran porcentaje de respuestas siendo within an hour, donde se ve drásticamente el cambio entre los anfitriones que responden en ese periodo de tiempo con el resto. Girona comparte un comportamiento muy similar a Ciudad de México, sin embargo en California no se tiene es cambio tan drástico. Se observa que está más balanceado los valores, por lo que podemos inferir que los anfitriones de California tardan más en contestarle a sus clientes. 

## host_acceptance_rate

```python
Cal['host_acceptance_rate'].value_counts().values
```

```python
plt.figure(figsize=(15,5))
plt.fill_between(Cal['host_acceptance_rate'].value_counts().sort_index(ascending=False).index,Cal['host_acceptance_rate'].value_counts().values,color='#22BBBB', alpha=0.7)
plt.title('CALIFORNIA HOST ACCEPTANCE RATE')
plt.xlabel('Acceptance Rate')
plt.xlim(0,100)
plt.show()

plt.figure(figsize=(15,5))
plt.fill_between(Mex['host_acceptance_rate'].value_counts().sort_index(ascending=False).index,Mex['host_acceptance_rate'].value_counts().values, color='#22BB13', alpha=0.7)
plt.title('MEXICO HOST ACCEPTANCE RATE')
plt.xlabel('Acceptance Rate')
plt.xlim(0,100)

plt.show()
plt.figure(figsize=(15,5))
plt.fill_between(Esp['host_acceptance_rate'].value_counts().sort_index(ascending=False).index,Esp['host_acceptance_rate'].value_counts().values, color='#BB2222', alpha=0.7)
plt.title('GIRONA HOST ACCEPTANCE RATE')
plt.xlabel('Acceptance Rate')
plt.xlim(0,100)

plt.show()
```

Estas gráficas representan el ratio de que tanto aceptan los anfitriones a sus clientes. En este caso observamos que Ciudad de México y Girona tienen un 'acceptance rate' más pegado al 100%, teniendo un área muy chica en la gráfica únicamente cubriendo la esquina inferior derecha. Mientras tanto, California abarca más pedazo de la gráfica. Se observa como la línea empieza a subir mucho desde antes. Con esto concluímos que los anfitriones de California aceptan menos seguido a sus clientes, con lo que podemos empezar a inferir algunas cosas como que son más cuidadosos de sus inmuebles, o son más críticos en cuanto a quién se le renta y a quién no. 

# number_of_reviews

```python
Mex['number_of_reviews'].plot.hist(color='#22BB13', label='Ciudad de México')
plt.legend()
plt.title('NÚMERO DE RESEÑAS EN CIUDAD DE MÉXICO')
plt.xlabel('Número de Reseñas')
plt.xlim(0,120)
plt.show()

Cal['number_of_reviews'].plot.hist(color='#22BBBB', label='California')
plt.legend()
plt.title('NÚMERO DE RESEÑAS EN CALIFORNIA')
plt.xlabel('Número de Reseñas')
plt.xlim(0,120)
plt.show()

Esp['number_of_reviews'].plot.hist(color='#BB2222', label='Girona')
plt.legend()
plt.title('NÚMERO DE RESEÑAS EN GIRONA')
plt.xlabel('Número de Reseñas')
plt.xlim(0,120)
plt.show()
```

Se aprecia cómo Girona no tiene muchas reseñas en sus inmuebles. El número de reseñas apenas escala a 35, mientras que en California y en Ciudad de México se llegan a conseguir hasta 100-120 reseñas. Esto es interesante pues California tiene muchos menos registros de inmuebles, sin embargo si logran superar a ambas ciudasdes en esta variable. Puede significar que la gente que renta inmuebles en Caliornia es muy crítica, o que la gente que alquila en Girona y Ciudad de México no les dan la suficiente importancia a esta parte del proceso. 

# review_scores_location 

```python
plt.figure(figsize=(15,5))
Mex['review_scores_location'].plot.hist(color='#22BB13', label='Ciudad de México')
plt.title('RESEÑAS DE UBICACIÓN CIUDAD DE MÉXICO')
plt.xlabel('Puntuación')
plt.legend()
plt.xlim(4,5)
plt.show()
plt.figure(figsize=(15,5))
Cal['review_scores_location'].plot.hist(color='#22BBBB', label='California')
plt.title('RESEÑAS DE UBICACIÓN CALIFORNIA')
plt.xlabel('Puntuación')
plt.legend()
plt.xlim(4,5)
plt.show()
plt.figure(figsize=(15,5))
Esp['review_scores_location'].plot.hist(color='#BB2222', label='Girona')
plt.title('RESEÑAS DE UBICACIÓN GIRONA')
plt.xlabel('Puntuación')
plt.legend()
plt.xlim(4,5)
plt.show()
```

A simple vista se observa como las reseñas entorno a la ubicación del inmueble llegan a ser menos placenteras en Girona, bajando hasta 4.3, mientras que en Ciudad de México y California lo mínimo que se obtiene es un 4.7. Cabe recalcar que 4.3 sigue siendo una buena calificación, sin embargo se puede inferir que los inmuebles de Ciudad de México y California disponibles en Airbnb están en zonas más bonitas, centradas y/o seguras.

#host_identity_verified

```python
Mex['host_identity_verified']
plt.pie(Mex['host_identity_verified'].value_counts().values[Mex['host_identity_verified'].value_counts().values>1], labels=['Verificado', 'No Verificado'], colors=['#22AA13','#00dd00'], startangle=90, shadow=True, explode=(0,0.13), autopct='%1.2f%%')
plt.title('ANFITRIÓN VERIFICADO CIUDAD DE MÉXICO')
plt.show()

Cal['host_identity_verified']
plt.pie(Cal['host_identity_verified'].value_counts().values, labels=['Verificado', 'No Verificado'], colors=['#22BBDD','#22BBaa'], startangle=90, shadow=True, explode=(0,0.13), autopct='%1.2f%%')
plt.title('ANFITRIÓN VERIFICADO CALIFORNIA')
plt.show()

Esp['host_identity_verified']
plt.pie(Esp['host_identity_verified'].value_counts().values[Esp['host_identity_verified'].value_counts().values>200], labels=['Verificado', 'No Verificado'], colors=['#992222','#dd2222'], startangle=90, shadow=True, explode=(0,0.13), autopct='%1.2f%%')
plt.title('ANFITRIÓN VERIFICADO GIRONA')
plt.show()
```

En estas gráficas de pastel se observa como Ciudad de México es la ciudad con más anfitriones verificados, es decir que tienen registrados en la plataforma datos personales como sus nombres, fecha de nacimiento, identificación oficial entre otros documentos. Esto es algo que causa seguridad para los inquilinos. Observamos como CIudad de México es superior en este aspecto, mientras que Girona y California tienen más de 11.5% de los anfitriones sin verificarse en la plataforma.
