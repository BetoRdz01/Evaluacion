Se observa que los precios ya no se elevan más allá de 10,000 pesos la noche, dado a que eran datos atípicos que se cambiaron con la media en la limpieza de los datos. Al utilizar rangos intercuartílicos como método de selección de valores atípicos, se es más estricto que con el método de desviación estándar, por lo que probablemente si utilizamos ese método si se alcanzaría a apreciar algunas opciones de casas en este rango de precios.
